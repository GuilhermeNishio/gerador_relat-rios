```python
# Função de criação de
def filtro(nome_distrito):
  filtro = dados[dados['Distritos'] == nome_distrito]
  return filtro
```

```python
# Função de criação de variável a ser inserida
def cria_var(filtroX, variável):
  variavel = '{:,.1f}'.format(float(filtroX[variável].iloc[0])).replace(',', ';').replace('.', ',').replace(';', '.').replace(',0', '')
  return variavel
```
